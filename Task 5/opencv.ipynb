```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# FIXED PATH (raw string + correct quotes)
image = cv2.imread(r'headphone_85x85.png')

if image is None:
    print("Image not found. Check path or filename.")
    exit()

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

scale_factor_1 = 3.0
scale_factor_2 = 1/3.0

height, width = image_rgb.shape[:2]

new_height = int(height * scale_factor_1)
new_width = int(width * scale_factor_1)
zoomed_image = cv2.resize(
    image_rgb, (new_width, new_height), interpolation=cv2.INTER_CUBIC
)

new_height1 = int(height * scale_factor_2)
new_width1 = int(width * scale_factor_2)
scaled_image = cv2.resize(
    image_rgb, (new_width1, new_height1), interpolation=cv2.INTER_AREA
)

fig, axs = plt.subplots(1, 3, figsize=(10, 4))
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image Shape: ' + str(image_rgb.shape))
axs[1].imshow(zoomed_image)
axs[1].set_title('Zoomed Image Shape: ' + str(zoomed_image.shape))
axs[2].imshow(scaled_image)
axs[2].set_title('Scaled Image Shape: ' + str(scaled_image.shape))

for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Correct path formatting
img = cv2.imread(r'headphone_85x85.png')

# Safety check
if img is None:
    print("Image not found. Check path or filename.")
    exit()

# Convert BGR to RGB for matplotlib
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Rotation parameters
center = (image_rgb.shape[1] // 2, image_rgb.shape[0] // 2)
angle = 30
scale = 1

# Get rotation matrix and rotate
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale)
rotated_image = cv2.warpAffine(image_rgb, rotation_matrix, (img.shape[1], img.shape[0]))

# Display using matplotlib
fig, axs = plt.subplots(1, 2, figsize=(7, 4))
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')
axs[1].imshow(rotated_image)
axs[1].set_title('Image Rotation')

for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Correct path using raw string
img = cv2.imread(r'headphone_85x85.png')

# Safety check
if img is None:
    print("Image not found. Check path or filename.")
    exit()

# Convert BGR to RGB for matplotlib
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
width, height = image_rgb.shape[1], image_rgb.shape[0]

# Translation parameters
tx, ty = 100, 70
translation_matrix = np.array([[1, 0, tx], [0, 1, ty]], dtype=np.float32)

# Apply translation
translated_image = cv2.warpAffine(image_rgb, translation_matrix, (width, height))

# Display
fig, axs = plt.subplots(1, 2, figsize=(7, 4))
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')
axs[1].imshow(translated_image)
axs[1].set_title('Image Translation')

for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread(r'headphone_85x85.png')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
kernel = np.ones((3, 3), np.uint8)

dilated = cv2.dilate(image_gray, kernel, iterations=2)
eroded = cv2.erode(image_gray, kernel, iterations=2)
opening = cv2.morphologyEx(image_gray, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(image_gray, cv2.MORPH_CLOSE, kernel)

fig, axs = plt.subplots(2, 2, figsize=(7, 7))
axs[0, 0].imshow(dilated, cmap='Greys'), axs[0, 0].set_title('Dilated Image')
axs[0, 1].imshow(eroded, cmap='Greys'), axs[0, 1].set_title('Eroded Image')
axs[1, 0].imshow(opening, cmap='Greys'), axs[1, 0].set_title('Opening')
axs[1, 1].imshow(closing, cmap='Greys'), axs[1, 1].set_title('Closing')

for ax in axs.flatten():
    ax.set_xticks([]), ax.set_yticks([])

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np

# Open the image with correct raw string path
img = cv2.imread(r'headphone_85x85.png')

# Safety check
if img is None:
    print("Image not found. Check path or filename.")
    exit()

# Apply log transform
c = 255 / (np.log(1 + np.min(img)))
log_transformed = c * np.log(1 + img)

# Convert to uint8
log_transformed = np.array(log_transformed, dtype=np.uint8)

# Save the output
cv2.imwrite('log_transformed.jpg', log_transformed)
print("Log transformed image saved as 'log_transformed.jpg'")
```

```python
import cv2
import numpy as np

# Function to map each intensity level to output intensity level.
def pixelVal(pix, r1, s1, r2, s2):
    if 0 <= pix <= r1:
        return (s1 / r1) * pix
    elif r1 < pix <= r2:
        return ((s2 - s1) / (r2 - r1)) * (pix - r1) + s1
    else:
        return ((255 - s2) / (255 - r2)) * (pix - r2) + s2

# Open the image (use correct path if needed)
img = cv2.imread(r'headphone_85x85.png')

# Safety check
if img is None:
    print("Image not found. Check path or filename.")
    exit()

# Define parameters
r1 = 70
s1 = 0
r2 = 140
s2 = 255

# Vectorize the function to apply it to each value in the Numpy array
pixelVal_vec = np.vectorize(pixelVal)

# Apply contrast stretching
contrast_stretched = pixelVal_vec(img, r1, s1, r2, s2)

# Convert to uint8
contrast_stretched = np.array(contrast_stretched, dtype=np.uint8)

# Save edited image
cv2.imwrite('contrast_stretch.jpg', contrast_stretched)
print("Contrast stretched image saved as 'contrast_stretch.jpg'")
```
